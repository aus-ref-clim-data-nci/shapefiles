## River regions

As summaried at the Bureau of Meteorology page on [Topographic Drainage Divisions and River Regions](http://www.bom.gov.au/water/about/riverBasinAuxNav.shtml), the Australian continent is divided into 12 drainage divisions that are further sub-divided into river regions.

```python
import geopandas as gp
from IPython.display import Image
import matplotlib.pyplot as plt
```

```python
#http://www.bom.gov.au/water/geofabric/documents/BOM002_Map_Poster_A3_Web.pdf
Image(filename='figures/river_regions.png', width=700)
```

```python
river_regions_file = 'V32_RiverRegions/V32_RiverRegions.shp'
```

```python
river_regions = gp.read_file(river_regions_file)
```

```python
river_regions
```

```python
river_regions.plot(
    figsize=[5, 5],
    column='Division',
    cmap='tab20',
    legend=True,
    linewidth=0.25,
    edgecolor='black',
    legend_kwds={'loc': 'right', 'bbox_to_anchor': (1.7, 0.5)},
)
plt.title('Drainage Divsions')
plt.xlabel('Longitude')
plt.ylabel('Latitude')
plt.show()
```

```python
river_regions[river_regions['Division'] == 'Tasmania'].plot(
    figsize=[5, 6],
    column='RivRegName',
    cmap='tab20',
    legend=True,
    linewidth=0.25,
    edgecolor='black',
    legend_kwds={'loc': 'right', 'bbox_to_anchor': (1.9, 0.5)},
)
plt.title('River Regions, Tasmania Drainage Division')
plt.xlabel('Longitude')
plt.ylabel('Latitude')
plt.show()
```
