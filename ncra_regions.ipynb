# NCRA regions

The Australian Climate Service (ACS)
defines a set of geographic regions defined for the National Climate Risk Assessment covering the mainland of Australia.

This notebook gives a basic demonstration of the shapefiles found at:  
`/g/data/ia39/aus-ref-clim-data-nci/shapefiles/data/ncra_regions/`

The shapefiles were downloaded from:
https://www.acs.gov.au/datasets/4f8b960fd3694fc28d6ff3d1278e9e75

```python
import geopandas as gp
from IPython.display import Image
import matplotlib.pyplot as plt
```

```python
ncra_file = '/g/data/ia39/aus-ref-clim-data-nci/shapefiles/data/ncra_regions/ncra_regions.shp'
```

```python
ncra_regions = gp.read_file(ncra_file)
ncra_regions
```

```python
ncra_regions.plot(
    figsize=[5, 5],
    column='label',
    legend=True,
    linewidth=0.5,
    edgecolor='black',
    legend_kwds={'loc': 'right', 'bbox_to_anchor': (1.5, 0.5)},
)
plt.title('NCRA Regions')
plt.xlabel('Longitude')
plt.ylabel('Latitude')
plt.show()
```
