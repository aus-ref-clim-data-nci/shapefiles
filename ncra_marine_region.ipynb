# NCRA marine region

The Australian Climate Service (ACS)
defines a set of geographic regions for the National Climate Risk Assessment covering Australia's marine regions.

This notebook gives a basic demonstration of the shapefiles found at:  
`/g/data/ia39/aus-ref-clim-data-nci/shapefiles/data/NCRA_Marine_region/`

ACS have the shapefiles here: https://www.acs.gov.au/datasets/d4991b87027b47218be493f907ecb8a0/explore

The shapefiles were downloaded from: https://services-ap1.arcgis.com/Xoz8Es66HpfM8jP9/arcgis/rest/services/NCRA_Marine_Region_with_metadata/FeatureServer/0/query?outFields=*&where=1%3D1&f=geojson


```python
import geopandas as gp
from IPython.display import Image
import matplotlib.pyplot as plt
```

```python
ncra_marine_file = '/g/data/ia39/aus-ref-clim-data-nci/shapefiles/data/NCRA_Marine_region/NCRA_Marine_region.shp'
```

```python
ncra_marine_regions = gp.read_file(ncra_marine_file)
ncra_marine_regions
```

```python
ncra_marine_regions.plot(
    figsize=[5, 5],
    legend=True,
    linewidth=0.5,
    edgecolor='black',
)
plt.title('NCRA Marine Region')
plt.xlabel('Longitude')
plt.ylabel('Latitude')
plt.show()
```
