## States and territories of Australia

The Australian Bureau of Statistics maintains the Australian Statistical Geography Standard (ASGS). The latest release is [Edition 3](https://www.abs.gov.au/statistics/standards/australian-statistical-geography-standard-asgs-edition-3).

In order to create a shapefile for Australian states and territories, this notebook relies on the "Australia - 2021 - Shapefile" downloaded from [here](https://www.abs.gov.au/statistics/standards/australian-statistical-geography-standard-asgs-edition-3/jul2021-jun2026/access-and-downloads/digital-boundary-files).

```python
import geopandas as gp
```

```python
filedir = '/g/data/xv83/dbi599/shapefiles'
shapefile = f'{filedir}/AUS_2021_AUST_SHP_GDA2020/AUS_2021_AUST_GDA2020.shp'
gdf = gp.read_file(shapefile)
```

```python
gdf
```

```python
gdf = gdf.drop([1])
gdf = gdf.drop(columns=['CHG_FLAG21', 'CHG_LBL21'])
gdf
```

```python
gdf.plot()
```

```python
gdf.to_file(f'{filedir}/australia/australia.shp')
```
