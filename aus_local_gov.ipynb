## Local Government Areas of Australia

This notebook gives a basic demonstration of the shapefiles found at:  
`/g/data/jt48/aus-ref-clim-data-nci/shapefiles/data/aus_local_gov/`

The Australian Bureau of Statistics (ABS) maintains the Australian Statistical Geography Standard (ASGS).
The latest release is
[Edition 3](https://www.abs.gov.au/statistics/standards/australian-statistical-geography-standard-asgs-edition-3),
which includes a "Local Government Areas - 2022 - Shapefile" that can be downloaded
[here](https://www.abs.gov.au/statistics/standards/australian-statistical-geography-standard-asgs-edition-3/jul2021-jun2026/access-and-downloads/digital-boundary-files).

In order to produce the shapefiles demonstrated below,
the `processing_scripts/process_local_gov.py` script applies some minor changes
(it removes some redundant rows and columns) to the original ABS shapefiles.

```python
import geopandas as gp
import matplotlib.pyplot as plt
```

```python
gdf = gp.read_file('/g/data/jt48/aus-ref-clim-data-nci/shapefiles/data/aus_local_gov/aus_local_gov.shp')
```

```python
gdf
```

```python
gdf.plot(figsize=[10, 10])
```

An example plot for just one state/territory...

```python
nt = gdf[gdf['STE_NAME21'] == 'Northern Territory']
```

```python
nt.plot(
    figsize=[9, 9],
    column='LGA_NAME22',
    cmap='tab20',
    legend=True,
    linewidth=0.5,
    edgecolor='black',
    legend_kwds={'loc': 'right', 'bbox_to_anchor': (1.9, 0.5)},
)
plt.xlabel('Longitude')
plt.ylabel('Latitude')
plt.title('NT LGAs')
plt.show()
```
