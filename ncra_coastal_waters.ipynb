# NCRA coastal waters 

The Australian Climate Service (ACS)
defines a set of geographic regions defined for the National Climate Risk Assessment for coastal waters of Australia.

This notebook gives a basic demonstration of the shapefiles found at:  
`/g/data/ia39/aus-ref-clim-data-nci/shapefiles/data/NCRA_regions_coastal_waters_GDA94/`


This dataset contains the shapefile of the National Climate Risk Assessment Geographic Regions, coastal waters. 

The base shapefile is from the Coastal Waters areas (AMB2020) dataset: [Coastal Waters areas (AMB2020) | Coastal Waters areas (AMB2020) | Australian Marine Spatial Information System (AMSIS) (arcgis.com)](https://amsis-geoscience-au.hub.arcgis.com/datasets/37a401e932544c88828a7d099880afb5_1/explore). 

The coastal waters region were extracted from the base dataset using the difference function (QGIS) and the NCRA regions (mainland) dataset: [NCRA regions | Australian Climate Service (acs.gov.au)](https://www.acs.gov.au/datasets/4f8b960fd3694fc28d6ff3d1278e9e75_0/about). 

The coastal waters of Queensland and Western Australia were split into north and south regions based on the coastal split in the NCRA regions (mainland) dataset.  The rationale for splitting WA and QLD into two regions is given in the published [National Climate Risk Assessment Methodology](https://www.dcceew.gov.au/climate-change/publications/national-climate-risk-assessment).

```python
import geopandas as gp
from IPython.display import Image
import matplotlib.pyplot as plt
```

```python
ncra_coastal_waters_file = '/g/data/ia39/aus-ref-clim-data-nci/shapefiles/data/NCRA_regions_coastal_waters_GDA94/NCRA_regions_coastal_waters_GDA94.shp'
```

```python
ncra_coastal_waters = gp.read_file(ncra_coastal_waters_file)
ncra_coastal_waters
```

```python
ncra_coastal_waters.plot(
    figsize=[5, 5],
    column='label',
    legend=True,
    linewidth=0.1,
    edgecolor='black',
    legend_kwds={'loc': 'right', 'bbox_to_anchor': (1.5, 0.5)},
)
plt.title('NCRA coastal waters')
plt.xlabel('Longitude')
plt.ylabel('Latitude')
plt.show()
```
